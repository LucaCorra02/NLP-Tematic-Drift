## 4️⃣ THEMATIC DRIFT SUMMARY DASHBOARD
Consolidated insights across all analyses

### 4.1 - Alignment vs Drift Correlation

```python
# Create composite view: Alignment scores + Topic lifecycle
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Mean alignment by year
df_alignment = plot_sim.df_merged.copy()
df_alignment['mean_alignment'] = df_alignment.filter(like='score_').mean(axis=1)
yearly_alignment = df_alignment.groupby('publication_year')['mean_alignment'].agg(['mean', 'std'])

axes[0,0].errorbar(yearly_alignment.index, yearly_alignment['mean'], 
                     yerr=yearly_alignment['std'], marker='o', capsize=5, label='Mean Alignment')
axes[0,0].set_title('Articles Alignment to Aims & Scope (per Year)', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Mean Alignment Score')
axes[0,0].grid(alpha=0.3)

# Plot 2: Topic lifecycle distribution
lifecycle_counts = drift_results['trend_classification'].value_counts()
axes[0,1].barh(lifecycle_counts.index, lifecycle_counts.values, color='steelblue')
axes[0,1].set_title('Topic Lifecycle Distribution', fontsize=12, fontweight='bold')
axes[0,1].set_xlabel('Number of Topics')

# Plot 3: Lexical vs Trend classification
cross_tab = pd.crosstab(drift_results['trend_classification'], drift_results['lexical_classification'])
cross_tab.plot(kind='bar', ax=axes[1,0], stacked=False)
axes[1,0].set_title('Trend vs Lexical Stability', fontsize=12, fontweight='bold')
axes[1,0].set_xlabel('Trend Classification')
axes[1,0].set_ylabel('Count')
axes[1,0].tick_params(axis='x', rotation=45)
axes[1,0].legend(title='Lexical Class', bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot 4: Lifecycle score distribution
axes[1,1].hist(drift_results['lifecycle_score'], bins=20, color='coral', edgecolor='black', alpha=0.7)
axes[1,1].set_title('Lifecycle Score Distribution', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('Lifecycle Score')
axes[1,1].set_ylabel('Frequency')
axes[1,1].axvline(drift_results['lifecycle_score'].mean(), color='red', linestyle='--', 
                   label=f'Mean: {drift_results["lifecycle_score"].mean():.3f}')
axes[1,1].legend()

plt.tight_layout()
plt.savefig('Similarity/dashboard_thematic_drift.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Thematic drift dashboard saved")
```

### 4.2 - Key Findings Summary

```python
print("\n" + "="*80)
print("📋 KEY FINDINGS - THEMATIC DRIFT ANALYSIS")
print("="*80)

print("\n1️⃣  ALIGNMENT WITH AIMS & SCOPE:")
print("-" * 80)
mean_align = df_alignment['mean_alignment'].mean()
aligned_pct = (df_alignment['mean_alignment'] > 0.5).sum() / len(df_alignment) * 100
drift_pct = (df_alignment['mean_alignment'] < 0.3).sum() / len(df_alignment) * 100

print(f"  • Overall mean alignment: {mean_align:.4f}")
print(f"  • Articles well-aligned (>0.5): {aligned_pct:.1f}%")
print(f"  • Articles showing drift (<0.3): {drift_pct:.1f}%")

print("\n2️⃣  TOPIC EVOLUTION PATTERNS:")
print("-" * 80)
emergent = len(drift_results[drift_results['trend_classification'] == 'Emergent'])
declining = len(drift_results[drift_results['trend_classification'] == 'Declining'])
stable = len(drift_results[drift_results['trend_classification'] == 'Stable'])
episodic = len(drift_results[drift_results['trend_classification'].str.contains('Episodic', na=False)])

print(f"  • Emergent topics: {emergent}")
print(f"  • Declining topics: {declining}")
print(f"  • Stable topics: {stable}")
print(f"  • Episodic topics: {episodic}")

print("\n3️⃣  LEXICAL STABILITY:")
print("-" * 80)
stable_keywords = len(drift_results[drift_results['lexical_classification'] == 'stable keywords'])
changing_keywords = len(drift_results[drift_results['lexical_classification'] == 'slowly changing keywords'])
different_keywords = len(drift_results[drift_results['lexical_classification'] == 'keywords are different'])

print(f"  • Topics with stable keywords: {stable_keywords}")
print(f"  • Topics with slowly changing keywords: {changing_keywords}")
print(f"  • Topics with different keywords (high drift): {different_keywords}")

print("\n4️⃣  TOP 5 MOST EMERGENT TOPICS:")
print("-" * 80)
top_emergent = drift_results[drift_results['trend_classification'] == 'Emergent'].nlargest(5, 'trend_Slope')[['topic_id', 'trend_Slope', 'burstiness', 'avg_Jaccard']]
print(top_emergent.to_string())

print("\n5️⃣  TOP 5 DECLINING TOPICS:")
print("-" * 80)
top_declining = drift_results[drift_results['trend_classification'] == 'Declining'].nsmallest(5, 'trend_Slope')[['topic_id', 'trend_Slope', 'burstiness', 'avg_Jaccard']]
print(top_declining.to_string())

print("\n" + "="*80)
```

```python
# Export comprehensive CSV report
export_cols = ['topic_id', 'trend_classification', 'lifecycle_score', 'lexical_classification',
               'burstiness', 'trend_p_val', 'trend_Slope', 'volatility_val', 'avg_Jaccard']

drift_results_export = drift_results[export_cols].sort_values('lifecycle_score', ascending=False)
drift_results_export.to_csv('Similarity/thematic_drift_complete_report.csv', index=False)

print("✅ Complete drift report exported to: Similarity/thematic_drift_complete_report.csv")
print(f"   Total topics analyzed: {len(drift_results_export)}")
```
